```python
import numpy as np
import cv2


def calc_histo(image, histSize, ranges=[0,256]):
    hist = np.zeros((histSize,1), np.float32)

    gap = ranges[1]/histSize

    for i in (image/gap).flat:
        hist[int(i)] += 1
    
    return hist


image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None : raise Exception("오류")

histSize, ranges = [32], [0,256]
gap = ranges[1]/histSize[0]

ranges_gap = np.arange(0, ranges[1]+1, gap)
hist1 = calc_histo(image, histSize[0], ranges)

hist2 = cv2.calcHist([image], [0], None, histSize, ranges)
hist3, bins = np.histogram(image,ranges_gap)

print(f"사용자 정의 함수 : \n {hist1.flatten()}")
print(f"cv 함수 : \n {hist2.flatten()}")
print(f"np 함수 : \n {hist3}")
```

```python
import numpy as np

def draw_histo(hist, shape=(200,256)):
    hist_img = np.full(shape, 255, np.uint8)
    cv2.normalize(hist,hist,0,shape[0], cv2.NORM_MINMAX)
    gap = hist_img.shape[1]/hist.shape[0]

    for i, h in enumerate(hist):
        x = int(round(i*gap))
        w = int(round(gap))
        cv2.rectangle(hist_img, (x,0,2,int(h)), 0, cv2.FILLED)

    return cv2.flip(hist_img,0)

image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None : raise Exception("오류")

hist = cv2.calcHist([image],[0],None,[32],[0,256])
hist_img = draw_histo(hist)

cv2.imshow("image",image)
cv2.imshow("hist_img", hist_img)
cv2.waitKey(0)
```

```python
import numpy as np
import cv2

def make_palette(rows):
    hue = [round(i*180/rows) for i in range(rows)]
    hsv = [ [[h,255,255]] for h in hue ]
    hsv = np.array(hsv, np.uint8)

    return cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)


def draw_hist_hue(hist, shape=(200,256,3)):
    hsv_palette = make_palette(hist.shape[0])
    hist_img = np.full(shape, 255, np.uint8)
    cv2.normalize(hist,hist,0,shape[0], cv2.NORM_MINMAX)
    gap = hist_img.shape[1]/hist.shape[0]

    for i, h in enumerate(hist):
        x = int(round(i*gap))
        w = int(round(gap))
        cv2.rectangle(hist_img, (x,0,2,int(h)), 0, cv2.FILLED)

    return cv2.flip(hist_img,0)


image = cv2.imread(r"img/img.jpg", cv2.IMREAD_COLOR)
if image is None : raise Exception("오류")

hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
hue_hist = cv2.calcHist([hsv_img],[0],None,[18],[0,180])
hue_hist_img = draw_hist_hue(hue_hist, (200,360,3))

cv2.imshow("image",image)
cv2.imshow("hist_img", hue_hist_img)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np

image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None : raise Exception("오류")

mean_5  = cv2.blur(image, (5,5))
mean_9  = cv2.blur(image, (9,9))

gauss_5 = cv2.GaussianBlur(image, (5,5), sigmaX=0)
gauss_9 = cv2.GaussianBlur(image, (9,9), sigmaX=0)

kernel_mean_5 = np.ones((5,5), np.float32)/25.0
f2d_mean_5 = cv2.filter2D(image, ddepth=-1, kernel=kernel_mean_5)

cv2.imshow("image", image)
cv2.imshow("mean_5", mean_5)
cv2.imshow("mean_9", mean_9)
cv2.imshow("gauss_5", gauss_5)
cv2.imshow("gauss_9", gauss_9)
cv2.imshow("f2d_mean_5", f2d_mean_5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None : raise Exception("오류")

data1 = [0,-1,0,
         -1,5,-1,
         0,-1,0]

data2 = [[-1,-1,-1],
         [-1,9,-1],
         [-1,-1,-1]]

mask1 = np.array(data1, np.float32).reshape(3,3)
mask2 = np.array(data2, np.float32)

sharpen1 = cv2.filter2D(image, ddepth=-1, kernel=mask1)
sharpen2 = cv2.filter2D(image, ddepth=-1, kernel=mask2)

cv2.imshow("image", image)
cv2.imshow("sharpen1", sharpen1)
cv2.imshow("sharpen2", sharpen2)
cv2.waitKey(0)

```
