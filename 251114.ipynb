```python
# 로버츠 에지 검출
import numpy as np
import cv2

def differential(image, data1, data2):
    mask1 = np.array(data1, np.float32).reshape(2,2)
    mask2 = np.array(data2, np.float32).reshape(2,2)

    dst1 = cv2.filter2D(image, ddepth=cv2.CV_32F,kernel=mask1)
    dst2 = cv2.filter2D(image, ddepth=cv2.CV_32F,kernel=mask2)
    dst = cv2.magnitude(dst1, dst2)
    dst1, dst2 = np.abs(dst1),np.abs(dst2)

    dst = np.clip(dst, 0, 255).astype("uint8")
    dst1 = np.clip(dst1, 0, 255).astype("uint8")
    dst2 = np.clip(dst2, 0, 255).astype("uint8")

    return dst, dst1, dst2

image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None: raise Exception("영상파일 읽기 오류")

data1 = [-1,0,
          0,1]
data2 = [0,-1,
          1,0]

dst, dst1, dst2 = differential(image, data1, data2)

cv2.imshow("image", image)
cv2.imshow("roberst edge", dst)
cv2.imshow("dst1", dst1)
cv2.imshow("dst2", dst2)
cv2.waitKey(0)
```

```python
# 라플라시안 에지 검출

import numpy as np
import cv2

image  = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None : raise Exception("영상파일 읽기 오류")

data1 = [
    [0,1,0],
    [1,-4,1],
    [0,1,0]
]

data2 = [
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
]

mask4 = np.array(data1, np.int16)
mask8 = np.array(data2, np.int16)

dst1 = cv2.filter2D(image, cv2.CV_16S, mask4)
dst2 = cv2.filter2D(image, cv2.CV_16S, mask8)

cv2.imshow("image", image)
cv2.imshow("filter2D 4-direction", cv2.convertScaleAbs(dst1))
cv2.imshow("filter2D 8-direction", cv2.convertScaleAbs(dst2))
cv2.waitKey(0)
```

```python
# 캐니 에지 검출
import numpy as np
import cv2

image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None: raise Exception("영상파일 읽기 오류") 

blur = cv2.GaussianBlur(image, (5,5), sigmaX=0)
canny_edges = cv2.Canny(blur,50,150)

sobel_gx = cv2.Sobel(image,cv2.CV_32F,1,0,ksize=3)
sobel_gy = cv2.Sobel(image,cv2.CV_32F,0,1,ksize=3)
sobel_mag = cv2.magnitude(sobel_gx, sobel_gy)
sobel_edges = cv2.convertScaleAbs(sobel_mag)

lap = cv2.Laplacian(image, cv2.CV_32F,ksize=3)
lap_edges = cv2.convertScaleAbs(lap)

cv2.imshow("image", image)
cv2.imshow("canny edges", canny_edges)
cv2.imshow("sobel edges", sobel_edges)
cv2.imshow("laplacian edges", lap_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 크기 변경
import numpy as np
import cv2
import time

def time_check(func):
    def wraaper(*args, **kwargs):
        start = time.perf_counter()
        ret_img = func(kwargs['image'], kwargs['size'])
        elapsed = (time.perf_counter()-start)*1000
        print(f"{kwargs['title']} 수행시간 = {elapsed:.2f}ms")
        return ret_img
    return wraaper


@time_check
def scaling(img, size):
    dst = np.zeros(size[::-1], img.dtype)
    ratioY, ratioX = np.divide(size[::-1], img.shape[:2])
    y = np.arange(0, img.shape[0], 1)
    x = np.arange(0, img.shape[1], 1)
    y, x = np.meshgrid(y,x)
    i, j = np.int32(y*ratioY), np.int32(x*ratioX)
    dst[i,j] = img[y,x]
    return dst

@time_check
def scaling2(img, size):
    dst = np.zeros(size[::-1], img.dtype)
    ratioY, ratioX = np.divide(size[::-1], img.shape[:2])
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            i,j = int(y*ratioY), int(x*ratioX)
            dst[i,j] = img[y,x]
    return dst

image = cv2.imread(r"img/img.jpg", cv2.IMREAD_GRAYSCALE)
if image is None: raise Exception("영상파일 읽기 오류")

W,H=image.shape[1], image.shape[0]
dst1 = scaling(image=image, size=(W//2,H//2), title="좌표행렬 방식(축소)->")
dst2 = scaling2(image=image, size=(W//2, H//2),title="for문 방식(축소)->")

dst3 = scaling(image=image, size=(W*2,H*2), title="좌표행렬 방식(확대)->")
dst4 = scaling2(image=image, size=(W*2, H*2),title="for문 방식(확대)->")

cv2.imshow("image", image)
cv2.imshow("dst1", dst1)
cv2.imshow("dst2", dst2)
cv2.imshow("dst3",dst3)
cv2.imshow("dst4", dst4)
cv2.waitKey(0)
```
